#### 1. Supongamos que x es un vector numérico. Explica en detalle, como las siguientes expresiones son evaluadas y que valores toman.

EXPRESION      | ¿QUE HACE?
:-------|------:
sum(!is.na(x)) | En primera instancia "!is.na(x)" devuelve TRUE o FALSE para cada elemento evaluado en el vector,TRUE si no es un valor perdido (NA) y FALSE si lo es,luego "sum" identifica estos resultados como ceros y unos,para luego sumarlos. En otras palabras esta expresion calcula el numero de elementos que no son NA.
c(x,x[-(1:length(x))])|En primera instancia "x[-(1:length(x))]" devuelve un vector,el cual resulta de quitarle a x los elementos que se encuentren en las posiciones comprendidas desde el 1 hasta length(x),pero vemos que todos los elementos de x cumplen esta condicion,por lo tanto al final se devolvera un vector sin elementos.En segunda instancia la expresion c(x,y) indexa dos vectores,por lo tanto para nuestro ejemplo,el vector "y" sera un vector sin elementos y la indexacion dara como resultado el mismo vector x.
x[length(x) + 1]/length(x)|"x[length(x) + 1]" devuelve el elemento de x que encuentre en la posicion "length(x) + 1" pero no existe ningun valor de x en ese posicion,ya que x solo poee length(x) elementos,luego devolvera el valor NA (elemento perdido o no encontrado).La division posterior simplemente seguira siendo NA.
sum(x > mean(x))|Esta expresion nos da como resultado el numero de elementos que que esten por encima de la media de x.

```r
#Aqui un ejemplo:
x=c(3,6,9,4,2)
sum(!is.na(x))
c(x,x[-(1:length(x))])
x[length(x) + 1]/length(x)
sum(x >mean(x))
```
